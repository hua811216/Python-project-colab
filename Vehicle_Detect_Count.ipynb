```
# Import libraries
from PIL import Image
import cv2
import numpy as np
```

```
# Reading image form url
image = cv2.imread('/content/drive/MyDrive/PythonData/cars.png')
type(image)
```

```
cc = cv2.CascadeClassifier('/content/drive/MyDrive/PythonData/cars.xml')
```

```
print(cv2.data.haarcascades + "/content/drive/MyDrive/datas/cars.xml")
```

```
image.shape
```

```
grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)
```

```
blur = cv2.GaussianBlur(grey,(5,5),0)
Image.fromarray(blur)
```

```
dilated = cv2.dilate(blur,np.ones((3,3)))
Image.fromarray(dilated)
```

```
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel) 
Image.fromarray(closing)
```

```
cars = cc.detectMultiScale(closing, scaleFactor=1.1, minNeighbors=1)
```

```
cnt = 0
for (x,y,w,h) in cars:
  cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
  cnt += 1
print(cnt, " cars found")
Image.fromarray(image)
```

```
# Reading image form url
image2 = cv2.imread('/content/drive/MyDrive/PythonData/bus2.jpg')
image3 = cv2.imread('/content/drive/MyDrive/PythonData/bus3.jpg')

grey2 = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
grey3 = cv2.cvtColor(image3,cv2.COLOR_BGR2GRAY)
```

```
bc = cv2.CascadeClassifier('/content/drive/MyDrive/PythonData/Bus_front.xml')
```

```
bus = bc.detectMultiScale(grey2, 1.1, 1)
bus3 = bc.detectMultiScale(grey3, 1.1, 1)
```

```
cnt = 0
for (x,y,w,h) in bus:
  cv2.rectangle(image2,(x,y),(x+w,y+h),(255,0,0),2)
  cnt += 1
print(cnt, " bus's found")
Image.fromarray(image2)
```

```
cnt = 0
for (x,y,w,h) in bus3:
  cv2.rectangle(image3,(x,y),(x+w,y+h),(255,0,0),2)
  cnt += 1
print(cnt, " bus's found")
Image.fromarray(image3)
```
